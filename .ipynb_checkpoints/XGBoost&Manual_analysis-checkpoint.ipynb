```python
##### LOAD model

bst = pkl.load(open('segment_predictor.pkl', 'rb'))
```

```python
### or TRAIN MODEL

### Get saved df
data_xgb = pd.read_pickle("unscaled_combined.pkl")
#data_xgb = pd.read_pickle("all_unscaled_combined.pkl")

data_xgb = data_xgb.drop(['predicted_value','true_value','errors'],axis=1)
data_xgb = data_xgb.sample(frac=1)    # Shuffle data
display(data_xgb)
```

```python
############# Binary classification on clustering results
x = data_xgb.drop('discriminated',axis=1)   
y = data_xgb['discriminated']
X_xgb_train, X_xgb_test, y_xgb_train, y_xgb_test = train_test_split(x.to_numpy(), y.to_numpy(),test_size=0.3)
# X_xgb_train, X_xgb_val, y_xgb_train, y_xgb_val = train_test_split(X_xgb_train, y_xgb_train,test_size=0.1)
dtrain = xgb.DMatrix(X_xgb_train, label=y_xgb_train)
dtest = xgb.DMatrix(X_xgb_test, label=y_xgb_test)
#dval = xgb.DMatrix(X_xgb_val, label=y_xgb_val)
# print(len(X_xgb_train),len(X_xgb_val),len(X_xgb_test))
```

```python
############# Binary classification on clustering results
runs = 30

avg_acc = []
avg_f1 = []
avg_auc = []
for i in range(runs):
    x = data_xgb.drop('discriminated',axis=1)   
    y = data_xgb['discriminated']
    X_xgb_train, X_xgb_test, y_xgb_train, y_xgb_test = train_test_split(x.to_numpy(), y.to_numpy(),test_size=0.3)
    # X_xgb_train, X_xgb_val, y_xgb_train, y_xgb_val = train_test_split(X_xgb_train, y_xgb_train,test_size=0.1)
    dtrain = xgb.DMatrix(X_xgb_train, label=y_xgb_train)
    dtest = xgb.DMatrix(X_xgb_test, label=y_xgb_test)
    #dval = xgb.DMatrix(X_xgb_val, label=y_xgb_val)
    # print(len(X_xgb_train),len(X_xgb_val),len(X_xgb_test))

    param = { 'eta': 1,'colsample_by_level':0.5, 'objective': 'binary:logistic'}
    #param = {'colsample_by_level':0.5, 'objective': 'binary:logistic'}

    # param['nthread'] = 4
    param['eval_metric'] = 'auc'
    evallist = [(dtest, 'eval'), (dtrain, 'train')]

    num_round = 50
    bst = xgb.train(param, dtrain, num_round, evallist, early_stopping_rounds=5)
    ypred = bst.predict(dtest)
    bin_predictions = np.around(ypred, decimals=0)
    f1 = metrics.f1_score(y_xgb_test, bin_predictions, average='binary')
    acc = metrics.accuracy_score(y_xgb_test, bin_predictions)
    fpr, tpr, thresholds = metrics.roc_curve(y_xgb_test, bin_predictions)#, pos_label=1)
    auc = metrics.auc(fpr, tpr)
    avg_acc.append(acc)
    avg_f1.append(f1)
    avg_auc.append(auc)
print('Mean of {} runs'.format(runs))
print('Accuracy of {}'.format(round( mean(avg_acc) ,3)))
print('F1 score of {}'.format(round(mean(avg_f1),3)))
print('AUC of {}'.format(round(mean(avg_auc),3)))

# Save last model
pkl.dump(bst, open('segment_predictor.pkl', 'wb'))
```

# Manual error analyis



```python
# Load df
unscaled_combined = pd.read_pickle('unscaled_combined.pkl')

### Initial dataframe   
manual_analysis = pd.DataFrame()
welch_statistic = []
p_value = []
pearson_r = []
pearson_p_value = []
features = []
means_high = []
means_low = []
r2_high = []
r2_low = []

for col in unscaled_combined.drop(['predicted_value','errors','true_value','discriminated'],axis=1).columns:
    ######## Continuous features: Regressie plot en Pearson's r correlation
    if len(unscaled_combined[col].unique()) > 15:
        pearson_corr = stats.pearsonr(unscaled_combined[col], unscaled_combined['errors'])
        pearson_r.append(pearson_corr[0])
        pearson_p_value.append(pearson_corr[1])
        fig, ax = plt.subplots()
        sns.regplot(unscaled_combined[col], unscaled_combined['errors'], ax=ax,scatter_kws={'s':1})
    
    
    ###### Devide categorical features into 2 groups by median of unique values, eg 4+ bedrooms vs 4- bedrooms, total_rooms 3.0+ vs total_rooms 3.0-
    if len(unscaled_combined[col].unique())<= 15:
        split_at = median(unscaled_combined[col].unique())
        pearson_r.append(None)
        pearson_p_value.append(None)
    ###### Split continous data into 2 groups by median
    else:
        split_at = median(unscaled_combined[col])
    if split_at == 0.5:
        high_group = '{} {}'.format(col, split_at+0.5)
        low_group = '{} {}'.format(col, split_at-0.5)
    else:
        high_group = '{} {}+'.format(col, split_at)
        low_group = '{} {}-'.format(col, split_at)
    unscaled_combined[high_group] = unscaled_combined[col] >= split_at
    unscaled_combined[low_group] = unscaled_combined[col] < split_at

    ##### Test the difference Welch T test
    welch_test = stats.ttest_ind(unscaled_combined[unscaled_combined[high_group]==True]['errors'], 
                                 unscaled_combined[unscaled_combined[low_group]==True]['errors'], 
                                 equal_var=False)
    means_high.append(np.mean(unscaled_combined[unscaled_combined[high_group]==True]['errors']))
    means_low.append(np.mean(unscaled_combined[unscaled_combined[low_group]==True]['errors']))

    fig, ax = plt.subplots()
    for group in [high_group, low_group]:         
#             mse = metrics.mean_squared_error(unscaled_combined[unscaled_combined[group]==True][['true_value']],
#                               unscaled_combined[unscaled_combined[group]==True][['predicted_value']])
        
        #r2 = hbac_kmeans.accuracy(unscaled_combined[unscaled_combined[group]==True])     
#         print('R2: ', round(r2,3))
#             print('For {}'.format(group) ) 
#             print('Mean abs relative error: ', np.mean(unscaled_combined[unscaled_combined[group]==True]['abs_error']))

#             print('MSE: ', round(mse,3))
        if group == high_group:
            r2_high.append(hbac_kmeans.accuracy(unscaled_combined[unscaled_combined[group]==True]))
            welch_statistic.append(welch_test[0])
            p_value.append(welch_test[1])
            features.append('{} vs {}'.format(high_group, low_group))
        else:
            r2_low.append(hbac_kmeans.accuracy(unscaled_combined[unscaled_combined[group]==True]))            
        sns.distplot(unscaled_combined[unscaled_combined[group]==True]['errors'], ax=ax)
        plt.legend(labels=['{}'.format(high_group),'{}'.format(low_group)])
    plt.show()
    
    ###### FOR boxplots
#         fig, ax = plt.subplots()
#         sns.boxplot(x=high_group, y="abs_error", hue=high_group, data=unscaled_combined,ax=ax)#, palette="PRGn")


manual_analysis['feature'] = features
manual_analysis['r2_group_high'] = r2_high
manual_analysis['r2_group_low'] = r2_low
manual_analysis['abs_r2_difference'] = abs(manual_analysis['r2_group_high']- manual_analysis['r2_group_low'])
manual_analysis['mean_error_group_high'] = means_high
manual_analysis['mean_error_group_low'] = means_low
manual_analysis['abs_difference'] = abs(manual_analysis['mean_error_group_high'] - manual_analysis['mean_error_group_low'])
manual_analysis['welch_statistic'] = welch_statistic
manual_analysis['p_value'] = p_value
manual_analysis['pearson_r'] = pearson_r
manual_analysis['pearson_p'] = pearson_p_value
manual_analysis = manual_analysis.round(decimals=3)
```

```python
## Sort on abs difference
manual_analysis = manual_analysis.sort_values(by='abs_difference',ascending=False)
# df_marked = manual_analysis.drop(['r2_group_high','r2_group_low'],axis=1).style.apply(lambda x: ['background: lightgreen' 
#                                   if (x['p_value']<0.05 or x['pearson_p_value']<0.05)
#                                   else '' for i in x], axis=1)

# with pd.option_context('display.max_rows', None, 'display.max_columns', None):
#     display(df_marked)

pd.set_option("max_colwidth", 100)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(manual_analysis.drop(['r2_group_high','r2_group_low','abs_r2_difference'],axis=1))

    
manual_analysis = manual_analysis.sort_values(by='abs_r2_difference',ascending=False)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(manual_analysis[['feature','r2_group_high','r2_group_low','abs_r2_difference']])

# ** INDEX (first col) = SORTED ON DIFFERNCE DISCRM. VS REMAINING CLUSTER!!!
```
